```python
Question 3.
def summarize_column(df):
    summary = {}
    for col in df.columns:
        summary[col] = {}
        summary[col]['Data Type'] = df[col].dtype
        summary[col]['Non-null Count'] = df[col].notnull().sum()
        summary[col]['Unique Values'] = df[col].nunique()
        if pd.api.types.is_numeric_dtype(df[col]):
            summary[col]['Mean'] = df[col].mean()
            summary[col]['Min'] = df[col].min()
            summary[col]['Max'] = df[col].max()
        elif pd.api.types.is_object_dtype(df[col]):
            summary[col]['Most Frequent'] = df[col].mode()[0]
    
    return pd.DataFrame(summary)

summaries = summarize_column(titanic_data)
print(summaries)
```

```python
Question 4.

If a dataset has non-numeric variables and missing values in numeric variables, df.describe() will only report the columns that have numerical values, as opposed to df.shape which will report every column.
As for the values it reports in the "count" column, it will "omit" the missing numerical values-in other words the count number of the column will decrease by the number of missing values found within that column.
In the example of the Titanic dataset, when the code df.shape is run, the output is (891, 15). However, when we run df.describe(), it only analyzes six of the fifteen columns, the ones that have numerical values. 
In the age column we can see it has a count value of 714 as opposed to 891 like the rest, since there are 177 values missing from the age column.
 
```
